```python
import os
import glob
from collections import Counter
import re

def compute_word_distribution(directory):
    word_counts = Counter()

    # Corrected variable name from 'directory_path' to 'directory'
    txt_files = glob.glob(os.path.join(directory, '*.txt'))

    for filepath in txt_files:
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read().lower()
            words = re.findall(r'\w+', content)
            word_counts.update(words)

    return word_counts

def write_top_k_words_to_file(directories, out_put_dir, title, k=50):
    output_filename = os.path.join(out_put_dir, "top_words.txt")

    with open(output_filename, 'w', encoding='utf-8') as file:
        file.write(title + "\n")
        file.write('-' * len(title) + "\n\n")  # Decorative underline

        for directory in directories:
            word_counts = compute_word_distribution(directory)
            file.write(f"Top {k} words from {directory}:\n")
            file.write("#index\t#word\t#frequency\n")
            for idx, (word, count) in enumerate(word_counts.most_common(k), 1):
                file.write(f"{idx}\t{word}\t{count}\n")
            file.write("\n\n")  # Add space between word lists from different directories

    print(f"Word frequencies have been written to {output_filename}\n")

```

```python
import os
import glob
from collections import Counter
import re

def compute_bigram_distribution(directory):
    bigram_counts = Counter()

    txt_files = glob.glob(os.path.join(directory, '*.txt'))

    for filepath in txt_files:
        with open(filepath, 'r', encoding='utf-8') as file:
            content = file.read().lower()
            words = re.findall(r'\w+', content)

            # Generate bigrams from the list of words
            bigrams = zip(words, words[1:])

            for bigram in bigrams:
                bigram_counts[bigram] += 1

    return bigram_counts

def write_top_k_bigrams_to_file(directories, out_put_dir, title, k=50):
    output_filename = os.path.join(out_put_dir, "top_bigrams.txt")

    with open(output_filename, 'w', encoding='utf-8') as file:
        file.write(title + "\n")
        file.write('-' * len(title) + "\n\n")  # Decorative underline

        for directory in directories:
            bigram_counts = compute_bigram_distribution(directory)
            file.write(f"Top {k} bigrams from {directory}:\n")
            file.write("#index\t#bigram\t#frequency\n")

            for idx, (bigram, count) in enumerate(bigram_counts.most_common(k), 1):
                # Join the bigram tuple into a string
                bigram_str = ' '.join(bigram)
                file.write(f"{idx}\t{bigram_str}\t{count}\n")
            file.write("\n\n")  # Add space between bigram lists from different directories

    print(f"Bigram frequencies have been written to {output_filename}\n")


```
